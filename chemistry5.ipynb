```python
targets=[]
Decoys=[]
FASTAFile().load("yeastFile.fasta",targets)
decoy=DecoyGenerator()
for entry in targets:
    rec=FASTAEntry(entry)
    rec.identifier="Decoy-"+rec.identifier
    seq=AASequence().fromString(rec.sequence)
    rec.sequence=decoy.reverseProtein(seq).toString()
    Decoys.append(rec)
```

```python
FASTAFile().store("yeastFileEditig.fasta",targets+Decoys)
```

```python
from Bio import SeqIO
target=[]
Decoy=[]

for seq_record in SeqIO.parse("yeastFile.fasta", "fasta"):
    target.append(seq_record)

```

```python
decoy=DecoyGenerator()

for entry in target:
    rev=entry
    rev.id="Decoy-"+rev.id
    string="".join(rev.seq.strip())
    seq = AASequence.fromString(string)
    rev.seq=decoy.reverseProtein(seq).toString()
    Decoy.append(rev)
    
```
